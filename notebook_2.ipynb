### Ejercicio 1.1 clasificador de temperatura
Escribir una función `clasificar_temperatura(t)` que devuelva un string según lo siguiente:
- Si t < 10 -> Frío
- Si 10 <= t < 20 -> Templado
- Si 20 <= t < 30 -> Cálido
- Si t >= 30 -> Calor

Ejemplos:
- `clasificar_temperatura(8)` -> 'Frío'
- `clasificar_temperatura(18)` -> 'Templado'
- `clasificar_temperatura(25)` -> 'Cálido'

```python
def clasificar_temperatura(t: float) -> str:
  if t < 10:
    return 'Frío'
  elif 10 <= t < 20:
    return 'Templado'
  elif 20 <= t < 30:
    return 'Cálido'
  else:
    return 'Calor'
```

### Ejercicio 1.2 - Tarifa de envío
Implementar costo_envio(peso_kg, express) con estas reglas:
- Base: hasta 1 kg cuesta 5
- Mas de 1 kg y hasta 5 cuesta 10
- Mas de 5kg cuesta 20
- Si express es True, sumar 7 adicionales

Ejemplos
- costo_envio(0.5, False) -> 5
- costo_envio(3, True) -> 17
- costo_envio(8, False) -> 20

```python
def costo_envio(peso_kg: float, express: bool = False) -> int:
  envio = 0
  if peso_kg < 0:
    print('Error, el peso tiene que ser mayor a cero')
  else:

    if peso_kg <= 1:
      envio = 5
    elif 1 < peso_kg <= 5:
      envio = 10
    else:
      envio = 20
  
    if express:
      envio += 7
    return envio
```

```python
costo_envio(-10)
```

```python
# Tests
assert costo_envio(0.5) == 5
assert costo_envio(3) == 10
assert costo_envio(6) == 20
assert costo_envio(0.5, express=True) == 12
assert costo_envio(3, express=True) == 17
assert costo_envio(6, express=True) == 27
print('OK - Ejercicios completados correctamente')
```

### Ejercicio 1.3 — Validación de contraseña

Creá `validar_password(p)` que devuelva:
- `'debil'` si tiene menos de 8 caracteres
- `'media'` si tiene al menos 8, pero **no** contiene números
- `'fuerte'` si tiene al menos 8 y contiene al menos un número

Pista: podés usar `any(ch.isdigit() for ch in p)`.

Ejemplos:
- `validar_password('hola')` → `'debil'`
- `validar_password('holahola')` → `'media'`
- `validar_password('holahola7')` → `'fuerte'`

```python
def validar_password(p: str) -> str:
  
  cant_caracteres = len(p)
  contiene_numeros = any(char.isdigit() for char in p)

  if cant_caracteres == 0:
    return 'Contraseña inválida'
  else:
    if cant_caracteres < 8:
      return 'debil'
    elif 8 <= cant_caracteres and not contiene_numeros:
      return 'media'
    else:
      return 'fuerte'
```

```python
validar_password('')
```

```python
# Tests
assert validar_password('hola') == 'debil'
assert validar_password('holahola') == 'media'
assert validar_password('holahola7') == 'fuerte'
print('OK - Ejercicio 1.3')
```

### Ejercicio 2.1 — Suma de pares

Completá `suma_pares(n)` para que devuelva la suma de todos los números pares entre 1 y `n` (incluyendo `n` si es par).

Ejemplos:
- `suma_pares(10)` → 30 (2+4+6+8+10)
- `suma_pares(1)` → 0

```python
def suma_pares(n:int) -> int:
  suma = 0
  for num in range(1, n+1):
    if num % 2 == 0:
      suma += num
  return suma
```

### Ejercicio 2.2 — Contar vocales

Implementá `contar_vocales(texto)` para contar cuántas vocales (a, e, i, o, u) aparecen en `texto`.
Debe ser case-insensitive.

Ejemplos:
- `contar_vocales('Hola Mundo')` → 4
- `contar_vocales('XYZ')` → 0

```python
def contar_vocales(texto: str) -> int:
  contador = 0
  vocales = 'aeiou'
  texto = texto.lower()
  
  if texto == '':
    print('El texto está vacío')
  else:
    for caracter in texto:
      if caracter in vocales:
        contador += 1
    return contador
```

```python
contar_vocales('')
```

```python

# Tests
assert contar_vocales('Hola Mundo') == 4
assert contar_vocales('XYZ') == 0
assert contar_vocales('AEIOU') == 5
print('OK - Ejercicio 2.2')
```

### Ejercicio 2.3 — Generar tabla de multiplicar

Completá `tabla_multiplicar(k, hasta)` que devuelva una lista con los múltiplos de `k` desde `k*1` hasta `k*hasta`.

Ejemplo:
- `tabla_multiplicar(3, 5)` → `[3, 6, 9, 12, 15]`

```python
def tabla_multiplicar(k:int, hasta:int) -> list:
    tabla = []
    for num in range(1, hasta + 1):
        tabla.append(k * num)
    return tabla
```

### Ejercicio 3.1 — Limpieza de lista

Dada una lista de strings `items`, eliminá todos los elementos vacíos (`''`) y devolvé la lista resultante.
Implementá `limpiar_items(items)` usando **métodos de lista**.

Ejemplo:
- `limpiar_items(['pan', '', 'leche', '', 'huevos'])` → `['pan', 'leche', 'huevos']`

```python
def limpiar_items(items: list[str]) -> list[str]:
    while '' in items:
        items.remove('')
    return items
```

```python
# Tests
assert limpiar_items(['pan', '', 'leche', '', 'huevos']) == ['pan', 'leche', 'huevos']
assert limpiar_items(['', '']) == []
assert limpiar_items(['ok']) == ['ok']
print('OK - Ejercicio 3.1')
```

### Ejercicio 3.2 — Insertar y ordenar

Implementá `insertar_y_ordenar(nums, x)` que:
1) Inserte `x` al final de la lista `nums`
2) Ordene la lista de menor a mayor **in place**
3) Devuelva la lista.

Ejemplo:
- `insertar_y_ordenar([3, 1, 2], 5)` → `[1, 2, 3, 5]`

```python
def insertar_y_ordenar(nums: list[int], x: int) -> list[int]:
    lista = nums.copy()
    lista.append(x)
    lista.sort()
    return lista
```

```python
# Tests
assert insertar_y_ordenar([3, 1, 2], 5) == [1, 2, 3, 5]
assert insertar_y_ordenar([], 7) == [7]
print('OK - Ejercicio 3.2')
```

### Ejercicio 3.3 — Stack (pila) con pop

Implementá `simular_pila(operaciones)` donde `operaciones` es una lista de strings del tipo:
- `'PUSH valor'` (por ejemplo `'PUSH 10'`)
- `'POP'`

La función debe usar una lista como pila:
- `PUSH` agrega al final
- `POP` quita del final (si la pila está vacía, ignorar)

Devolvé el contenido final de la pila.

Ejemplo:
- `simular_pila(['PUSH 5', 'PUSH 7', 'POP', 'PUSH 1'])` → `[5, 1]`

```python
def simular_pila(operaciones: list[str]) -> list[int]:
    pila: list[int] = []
    for comando in operaciones:
        operacion = comando.split()
        if operacion[0] == 'PUSH':
            pila.append(int(operacion[1]))
        elif operacion[0] == 'POP' and len(pila) > 0:
            pila.pop()
        else:
            print('Comando no reconocido')
    return pila
```

```python
# Tests
assert simular_pila(['PUSH 5', 'PUSH 7', 'POP', 'PUSH 1']) == [5, 1]
assert simular_pila(['POP', 'POP']) == []
assert simular_pila(['PUSH 3', 'PUSH 4']) == [3, 4]
print('OK - Ejercicio 3.3')
```

### Ejercicio 4.1 — Normalizar nombre

Implementá `normalizar_nombre(nombre)` que:
- Quite espacios al inicio y al final
- Ponga todo en minúscula
- Luego ponga la primera letra de cada palabra en mayúscula (Title Case)

Ejemplo:
- `normalizar_nombre('  jUaN pErEz  ')` → `'Juan Perez'`

```python
def normalizar_nombre(nombre: str) -> str:
  if nombre == '':
    print("El nombre está vacío")
  else:
    nombre = nombre.strip()
    nombre = nombre.lower()
    nombre = nombre.title()
    return nombre
```

```python
# Tests
assert normalizar_nombre('  jUaN pErEz  ') == 'Juan Perez'
assert normalizar_nombre('mARIA') == 'Maria'
print('OK - Ejercicio 4.1')
```

### Ejercicio 4.2 — Contar palabras

Completá `contar_palabras(frase)` para contar cuántas palabras hay.
Considerá que puede haber múltiples espacios. Usá métodos de string.

Ejemplos:
- `contar_palabras('hola mundo')` → 2
- `contar_palabras('  hola   mundo   cruel ')` → 3

```python
def contar_palabras(frase: str) -> int:
    frase = [item.strip() for item in frase.split()]
    cant_palabras = len(frase)
    return cant_palabras
```

### Ejercicio 4.3 — Sanitizar texto

Implementá `sanitizar(texto)` que:
- Reemplace comas `,` por espacios
- Reemplace puntos `.` por espacios
- Convierta a minúsculas
- Devuelva una lista de tokens (palabras) sin vacíos

Ejemplo:
- `sanitizar('Hola, Mundo. Python!')` → `['hola', 'mundo', 'python!']`

```python
def sanitizar(texto: str) -> list[str]:
    if texto == '':
        print("El texto está vacío")
    else:
        texto = texto.replace(',', ' ')
        texto = texto.replace('.', ' ')
        texto = texto.lower()
        texto = texto.split()
        return texto
```

```python
sanitizar('Uno,dos,tres.')
```

```python
# Tests
assert sanitizar('Hola, Mundo. Python!') == ['hola', 'mundo', 'python!']
assert sanitizar('Uno,dos,tres.') == ['uno', 'dos', 'tres']
print('OK - Ejercicio 4.3')
```

### Ejercicio 5.1 — Área de un círculo

Implementá `area_circulo(r)` que devuelva el área usando `math.pi`.

Ejemplo:
- `area_circulo(1)` → ~3.14159

Usá redondeo en los tests para evitar problemas de flotantes.

```python
import math
# pi = math.pi
def area_circulo(r: float) -> float:
  pi = math.pi
  if r <= 0:
    print("Error, el radio no puede ser negativo ni puede ser cero")
  else:  
    area = pi * (r**2)
    return area
```

### Ejercicio 5.2 — Función reutilizable: aplicar descuento

Creá `aplicar_descuento(precio, porcentaje)` que devuelva el precio final.
Luego, completá `precio_final_con_impuesto(precio, porcentaje_desc, impuesto)` que:
1) Aplique descuento usando `aplicar_descuento`
2) Sume impuesto (por ejemplo 0.21 para 21%)

Ejemplo:
- `precio_final_con_impuesto(100, 10, 0.21)` → 108.9

```python
def aplicar_descuento(precio: float, porcentaje: float) -> float:
    if precio <= 0 or porcentaje < 0 or porcentaje > 100:
        print("Error, el precio no puede ser negativo ni igual a 0 y el porcentaje debe estar entre 0 y 100")
    else:
        precio_total = precio - (precio * porcentaje / 100)
        return precio_total

def precio_final_con_impuesto(precio: float, porcentaje_desc: float, impuesto: float) -> float:
    precio_con_impuesto = aplicar_descuento(precio, porcentaje_desc) * (1 + impuesto)
    return precio_con_impuesto
```

### Ejercicio 5.3 — Funciones + listas: filtrar y transformar

Implementá `procesar_numeros(nums)` que:
- Filtre solo los números positivos
- Calcule el cuadrado de cada positivo
- Devuelva la nueva lista

Restricción: hacelo con una función y un `for` (sin list comprehensions en este ejercicio).

Ejemplo:
- `procesar_numeros([-2, 3, 0, 4])` → `[9, 16]`

```python
def procesar_numeros(nums: list[int]) -> list[int]:
  positivos = []
  for numero in nums:
    if numero > 0:
      positivos.append(numero**2)
  return positivos
```

```python
# Tests
assert procesar_numeros([-2, 3, 0, 4]) == [9, 16]
assert procesar_numeros([0, -1, -2]) == []
assert procesar_numeros([5]) == [25]
print('OK - Ejercicio 5.3')
```
